```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/1000_Companies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
sns.pairplot(df)
```

```python
plt.scatter(df['R&D Spend'],df['Profit'])
```

```python
plt.scatter(df['Marketing Spend'],df['R&D Spend'])
```

```python
x=df['R&D Spend'].values.reshape(-1,1)
y=df['Marketing Spend']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[153441.51]])
```

```python
y_pred = model.predict(x)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(x,y_pred)*100
```

```python
mean_squared_error(x,y_pred)*100
```

```python
plt.scatter(df['R&D Spend'],df['Marketing Spend'])
plt.plot(df['R&D Spend'],y_pred,color='r')
plt.show()
```
